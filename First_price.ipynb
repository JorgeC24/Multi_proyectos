```python
from dash import Dash, html, dcc, Input, Output

import yfinance as yf
```

```python
app = Dash()
```

```python
#ejemplo 1
app.layout = html.Div([
    #html.H1("Modelo de Markowitz"),
    dcc.Input(id="tickers", type = "text", value = "Escriba los tickers a analizar"),
    #html.Button("Calcular", id="submit", n_clicks=0),
    html.Div(children = '', id = 'output-text')
])

@app.callback(
    Output(component_id = 'output-text' , component_property = 'children'),
    Input(component_id = 'tickers', component_property = 'value')
)

def valor (tickers):
    #info = "" # Initialize the info variable
    
    data = yf.download(tickers, start="2020-01-01", end="2021-01-01")["Close"].dropna()
    
    return data.iloc[0]
```

```python
if __name__ == '__main__':
    app.run(debug = True)
```

```python


def valor (ticker):
    info = "" # Initialize the info variable
    
    data = yf.download(ticker, start="2020-01-01", end="2021-01-01")["Close"].dropna()
    
    for ticker_name in ticker: # Iterate through the list of ticker names
        # Get the first closing price for the current ticker and format it to two decimal places
        first_close_price = data[ticker_name].iloc[0]
        info += f"{ticker_name}: {first_close_price:.2f}<br>"

    return info

valor(["AAPL"])
```
